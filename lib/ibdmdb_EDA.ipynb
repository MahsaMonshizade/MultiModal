### Metadata

```python
import pandas as pd
```

```python
metadata = pd.read_csv('../Dataset/ibdmdb_data/metadata.csv')
```

```python
metadata.columns
```

```python
unique_values = metadata['External ID'].nunique()
```

```python
#number of samples
unique_values
```

```python
metadata.shape
```

```python
metadata
```

```python
participant_dict = {}
for index, row in metadata.iterrows():
    participant = row['Participant ID']
    week = row['week_num']
    
    # Check if participant already in dictionary
    if participant in participant_dict:
        participant_dict[participant].append(week)
    else:
        # If participant is not in dictionary, create a new entry
        participant_dict[participant] = [week]

print(participant_dict)
```

```python
# number of subjects
len(participant_dict)
```

```python
for k in participant_dict:
    participant_dict[k] = sorted(list(set(participant_dict[k])))
```

```python
for k in participant_dict:
    print(participant_dict[k])
```

```python
list(set(metadata['site_name'].values.tolist()))
```
